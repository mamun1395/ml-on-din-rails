```python
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
DATASET_PATH = './dataset/raw_dataset/'
IMAGE_PATH = './images/'
```

# Network Configuration 1

```python
# accumulating all datasets
```

```python
nc1_ncn5121_no4_files = glob.glob(DATASET_PATH + 'NetworkConfig1/ncn5121_no4/*')
nc1_ncn5121_no5_files = glob.glob(DATASET_PATH + 'NetworkConfig1/ncn5121_no5/*')
nc1_nodevice_files = glob.glob(DATASET_PATH + 'NetworkConfig1/nodevice/*')
nc1_switch_dem_files = glob.glob(DATASET_PATH + 'NetworkConfig1/switch_dem/*')
nc1_switch_sl2_files = glob.glob(DATASET_PATH + 'NetworkConfig1/switch_sl2/*')
```

```python
# reading datasets and taking mean of the datasets
```

```python
nc1_ncn5121_no4_data = pd.concat([pd.read_csv(file, header=None) for file in nc1_ncn5121_no4_files]).groupby([0], as_index=False).mean()
nc1_ncn5121_no5_data = pd.concat([pd.read_csv(file, header=None) for file in nc1_ncn5121_no5_files]).groupby([0], as_index=False).mean()
nc1_nodevice_data = pd.concat([pd.read_csv(file, header=None) for file in nc1_nodevice_files]).groupby([0], as_index=False).mean()
nc1_switch_dem_data = pd.concat([pd.read_csv(file, header=None) for file in nc1_switch_dem_files]).groupby([0], as_index=False).mean()
nc1_switch_sl2_data = pd.concat([pd.read_csv(file, header=None) for file in nc1_switch_sl2_files]).groupby([0], as_index=False).mean()
```

```python
# creating directory for saving plotted image files
```

```python
if not os.path.exists(IMAGE_PATH):
    os.mkdir(IMAGE_PATH)
```

```python
# plotting datasets
```

```python
fig_nc1 = plt.figure(figsize=(20, 10))
ax_nc1 = fig_nc1.add_subplot(111)
ax_nc1.plot(nc1_nodevice_data[0]/1000, 20*np.log10(nc1_nodevice_data[1]), label='nodevice')
ax_nc1.plot(nc1_ncn5121_no5_data[0]/1000, 20*np.log10(nc1_ncn5121_no5_data[1]), label='ncn5121_no5')
ax_nc1.plot(nc1_ncn5121_no4_data[0]/1000, 20*np.log10(nc1_ncn5121_no4_data[1]), label='ncn5121_no4')
ax_nc1.plot(nc1_switch_dem_data[0]/1000, 20*np.log10(nc1_switch_dem_data[1]), label='switch_dem')
ax_nc1.plot(nc1_switch_sl2_data[0]/1000, 20*np.log10(nc1_switch_sl2_data[1]), label='switch_sl2', linestyle='dashed', color='black')
ax_nc1.legend()
ax_nc1.set_title('Network Configuration 1')
ax_nc1.set_xlabel('Frequency (MHz)')
ax_nc1.set_ylabel('Gain (dB)')
ax_nc1.set_xlim(0, 25)
ax_nc1.set_ylim(-65, -25)
plt.savefig(IMAGE_PATH + 'network_config_1.pdf', bbox_inches='tight')
```
